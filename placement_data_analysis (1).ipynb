```python
import zipfile
import os

# Extract
with zipfile.ZipFile("/content/archive (4).zip", 'r') as zip_ref:
    zip_ref.extractall("dataset_folder")

# Check files
print(os.listdir("dataset_folder"))
```

```python
import pandas as pd

df = pd.read_csv("/content/archive (4).zip")  # use correct filename
df.head()
```

```python
df.info()
df.columns
df.describe()
```

```python
df.isnull().sum()
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Rename columns (optional)
df.rename(columns={'CGPA': 'CGPA'}, inplace=True)
```

```python
print(df.columns.tolist())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Placement', data=df)
plt.title("Placement Count")
plt.show()
```

```python
sns.countplot(x='Internship_Experience', hue='Placement', data=df)
plt.title("Internship Experience vs Placement")
plt.show()
```

```python
sns.boxplot(x='Placement', y='Projects_Completed', data=df)
plt.title("Projects Completed vs Placement")
plt.show()
```
